```python
import pandas as pd
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlations = data.corr()
correlations
```

```python
import seaborn as sns
sns.heatmap(correlations, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),
          y = data['mpg'])
print('Hệ số intercept là:', model.intercept_)
print('Hệ số coeficient là:',model.coef_)
```

```python
# độ chính xác
model.score(pd.DataFrame(data['wt']), data['mpg'])
```

```python
# dự đoán
y_pred=model.predict(X=pd.DataFrame(data['wt']))
y_pred
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(data['wt'],y_pred, color='violet')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T #Chuyển hàng thành cột
model.fit(X, y = data['mpg'])
print('Intercept của mô hình:', model.intercept_)
print('Coefficient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X,y=data['mpg'] ))
```

```python
import numpy as np
range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
import numpy as np
import pandas as pd
url = '/content/Advertising.csv'
data = pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlations = data.corr()
correlations
```

```python
import seaborn as sns
sns.heatmap(correlations, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(pd.DataFrame(data['TV']), data['sales'])
print('Intercept of the model:', model.intercept_)
print('Coeficient of the model:', model.coef_)
print('Accuracy of the model:', model.score(pd.DataFrame(data['TV']), data['sales']))
```

```python
y_pred = model.predict(pd.DataFrame(data['TV']))
print(y_pred)
```

```python
data.plot(kind='scatter', x='TV', y='sales', figsize=(9,9), color='black')
plt.plot(data['TV'], y_pred, color='violet')
```

```python
model = linear_model.LinearRegression()
model.fit(pd.DataFrame([data['TV'], data['TV']**2]).T, data['sales'])
print('Intercept:', model.intercept_)
print('Coeficient:', model.coef_)
print('Accuracy:', model.score(pd.DataFrame([data['TV'], data['TV']**2]).T, data['sales']))
```

```python
range = np.arange(0,320,0.1)
x_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(x_range)
data.plot(kind='scatter',
          x='TV',
          y='sales',
          figsize=(9,9),
          color='black')
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'], data['TV']**2, data['TV']**3]).T
model.fit(X, data['sales'])
print('Intercept:', model.intercept_)
print('Coeficient:', model.coef_)
print('Accuracy:', model.score(X, data['sales']))

range = np.arange(0,320,0.1)
x_range = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(x_range)
data.plot(kind='scatter',
          x='TV',
          y='sales',
          figsize=(9,9),
          color='black')
plt.plot(range,y_pred,color='green')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'], data['TV']**2, data['TV']**3, data['TV']**4, data['TV']**5]).T
model.fit(X, data['sales'])
print('Intercept:', model.intercept_)
print('Coeficient:', model.coef_)
print('Accuracy:', model.score(X, data['sales']))

range = np.arange(0,320,0.1)
x_range = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(x_range)
data.plot(kind='scatter',
          x='TV',
          y='sales',
          figsize=(9,9),
          color='black')
plt.plot(range,y_pred,color='green')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
url = '/content/iris.csv'
data = pd.read_csv(url)
data.head()
data.hist()
plt.show()
```

```python
correlations = data.corr()
correlations
```

```python
import seaborn as sns
sns.heatmap(correlations, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
x = pd.DataFrame(data['petal.width'])
model.fit(x, data['petal.length'])
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)
print('Accuracy:', model.score(x, data['petal.length']))
```

```python
y_pred = model.predict(x)
data.plot(kind='scatter', x='petal.width', y='petal.length', figsize=(9,9), color='black')
plt.plot(data['petal.width'], y_pred, color='violet')
```

```python
# 3nd order
x = pd.DataFrame([data['petal.width'], data['petal.width']**2, data['petal.width']**3]).T
model.fit(x, data['petal.length'])
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)
print('Accuracy:', model.score(x, data['petal.length']))
```

```python
import numpy as np
range = np.arange(0,2.7,0.1)
x_range = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(x_range)
data.plot(kind='scatter', x='petal.width', y='petal.length', figsize=(9,9), color='black')
plt.plot(range, y_pred, color='green')
```
